```python
import sys
sys.path.append('../')

import PIL.Image
from torchvision.transforms.transforms import ToTensor
import torch
from tools.img import tensor2ndarray
import matplotlib.pyplot as plt
```

```python
x_c = PIL.Image.open('../resource/example/000107.jpg')
scale = 224
x_c = x_c.resize((scale, scale))
x_c = ToTensor()(x_c)

x_fft = torch.fft.fft2(x_c, dim=(1, 2))
x_ifft = torch.fft.ifft2(x_fft, dim=(1, 2))
x_c_ = torch.real(x_ifft)

_, ax = plt.subplots(1, 2)
ax[0].imshow(tensor2ndarray(x_c))
ax[1].imshow(tensor2ndarray(x_c_))

plt.show()
```
